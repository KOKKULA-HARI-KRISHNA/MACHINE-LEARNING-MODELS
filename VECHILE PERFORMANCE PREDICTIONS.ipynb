# MODEL - 1 LINEAR REGRESSION WITH LOG

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
data_act = sm.add_constant(data_act)
```

```python
x1 = data_act.drop(['Kilometer_per_liter','Kilometer_per_liter_log'], axis = 1)
y1 = pd.DataFrame(data_act['Kilometer_per_liter_log'])
```

```python
x1_train,x1_test,y1_train,y1_test = train_test_split(x1,y1, test_size = 0.3, random_state = 1)
print(x1_train.shape)
print(x1_test.shape)
print(y1_train.shape)
print(y1_test.shape)
```

```python
LR_MODEL_WITH_LOG = sm.OLS(y1_train, x1_train).fit()
print(LR_MODEL_WITH_LOG.summary())
```

```python
y_predict_log = LR_MODEL_WITH_LOG.predict(x1_test)
y_predict = np.exp(y_predict_log)
y_actual = y1_test['Kilometer_per_liter_log']
```

```python
y_predict.shape
```

```python
y_actual.shape
```

```python
plt.scatter(y_predict,y_actual)
```

```python
R_SQUARE = LR_MODEL_WITH_LOG.rsquared
ADJ_SQUARE = LR_MODEL_WITH_LOG.rsquared_adj
AIC = LR_MODEL_WITH_LOG.aic
BIC = LR_MODEL_WITH_LOG.bic
```

```python
from statsmodels.tools.eval_measures import rmse
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
RMSE = rmse(y_predict,y_actual)
MAE = mean_absolute_error(y_predict,y_actual)
```

```python
MODEL_1 = pd.DataFrame({'MODELS':['LR_MODEL_WITH_LOG'],
                        'RMSE':[RMSE],
                        'MAE':[MAE],
                        'R_SQUARE':[R_SQUARE],
                        'ADJ_SQUARE':[ADJ_SQUARE],
                       'AIC':[AIC],
                       'BIC':[BIC]})
```

```python
MODEL_1
```
